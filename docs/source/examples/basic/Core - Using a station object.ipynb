```python
import prettymaps
from bootsoff.core import Station
```

## Create a station

```python
s01 = Station(label='S01', coords=(171000, 122000, 100), crs='EPSG:31370')
```

```python
s01.plot(show_label=True)
```

## Change CRS

```python
s01.to_crs(crs='EPSG:4326')
```
