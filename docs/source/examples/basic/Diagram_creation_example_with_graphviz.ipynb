# Example of diagram creation using the dot language
see i.e. https://en.wikipedia.org/wiki/DOT_(graph_description_language)  
O. Kaufmann, 2020

```python
from IPython.display import display
import graphviz
```

```python
dot = graphviz.Digraph(comment='Processing graph example')
dot.engine = 'dot' # choose between circo, dot, fdp, sfdp, neato, osage twopi or patchwork

dot.attr('node', shape='circle')
dot.node('0', 'Start')

dot.attr('node', shape='box')
with dot.subgraph() as s:
    s.attr(rank='same')
    s.node('A', 'Load GPS dataset')
    s.node('B', 'Load EM dataset')
dot.node('C', 'Filter data')
dot.node('D', 'Merge GPS and EM datasets')
dot.node('E', 'Plot map')

dot.attr('node', shape='circle')
dot.node('1', 'End')

dot.edges(['0A', '0B', 'AD', 'BC', 'CD', 'DE', 'E1'])
#dot.edge('B', 'L', constraint='false')
display(dot)
```

### Save graph as dot and svg files (that could be edited with vector graphics drawing software e.g. Inkscape )

```python
dot.format = 'svg'
dot.render(filename='output/example_graph.dot')
```

### Save graph as dot and png files

```python
dot.format = 'png'
dot.render(filename='output/example_graph.dot')
```
